```python
import requests
from bs4 import BeautifulSoup
from html_table_parser import parser_functions as parser
```

```python
names = []
for i in range(1, 1271):
    print(i)
    url= f'https://www.namechart.kr/chart/all?gender=t&page={i}'
    result = requests.get(url)
    soup = BeautifulSoup(result.text, 'html.parser')
    # data = soup.findAll("span", {"class":"font-semibold flex items-baseline"})
    elements = soup.select('span[class*="font-semibold flex items-baseline"]')
    for j in range(len(elements)):
        names.append(elements[j].text)
        
```

```python
import pickle

names = list(set(names))
with open('names.pickle', 'wb') as f:
    pickle.dump(names, f)
```
